```python
# Missing Values/Null Values Count
print(hotel.isnull().sum().sort_values(ascending=False))
```

##1.1 In this dataset we get 112593 null value in 'company' column so we have replaced the null values with 0

```python
#replace the null values 0.0 float
hotel['company'] = hotel['company'].fillna(0.0)
hotel['company'].head(20)
```

##1.2 In the 'agent' column we got 16340 null values so we have replaced it with 0


```python
#here also the value replace with 0.0
hotel['agent'] = hotel['agent'].fillna(0.0)
```

##1.3 In country we can not set the country values as 0 becuase it contains country codes representing different countries. So instead of replacing it with zero we have replaced it with the mode value in country column. Mode is nothing but just the most repeating value .

```python
hotel1=hotel.copy()
```

```python
hotel1.country.mode().to_string()
```

```python
hotel['country'].fillna(hotel1.country.mode().to_string(), inplace=True)
```

##1.4 Children Column has the count of children then we replaced missing values with mean values

```python
hotel['children'].fillna(round(hotel.children.mean()), inplace=True)
```

##1.5 In the dataset some rows have zero guests including adults,children and babies  

```python
hotel[(hotel.adults+hotel.babies+hotel.children)==0].shape
```


Many rows in the dataset contains values where there is no adult and only children and babies that makes no sense so we directly deleted it by using drop  

```python
hotel = hotel.drop(hotel[(hotel.adults+hotel.babies+hotel.children)==0].index)
```

So after doing all these cleaning let's check the dataset

```python
# Missing Values/Null Values Count
print(hotel.isnull().sum().sort_values(ascending=False))
```

```python
hotel.dtypes
```

```python
hotel[['children', 'company', 'agent']] = hotel[['children', 'company', 'agent']].astype('int64')
```

####Chart - 1 - Hotels type ratio

#### Chart - 10 - Hostogram and Boxplot

```python
# Chart - 9 visualization code
# Visualizing code of hist plot for each columns to know the data distibution
for col in hotel.describe().columns:
  fig=plt.figure(figsize=(9,6))
  ax=fig.gca()
  feature= (hotel[col])
  sns.histplot(hotel[col])
  ax.axvline(feature.mean(),color='magenta', linestyle='dashed', linewidth=2)
  ax.axvline(feature.median(),color='cyan', linestyle='dashed', linewidth=2)
  ax.set_title(col)
plt.show()

# Visualizing code of box plot for each columns to know the data distibution
for col in hotel.describe().columns:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    hotel.boxplot( col, ax = ax)
    ax.set_title('Label by ' + col)
    #ax.set_ylabel("Churn")
plt.show()
```

##### 1. Why did you pick the specific chart?

* The histogram is a popular graphing tool. It is used to summarize discrete or continuous data that are measured on an interval scale. It is often used to illustrate the major features of the distribution of the data in a convenient form. It is also useful when dealing with large data sets (greater than 100 observations). It can help detect any unusual observations (outliers) or any gaps in the data.

 Thus, I used the histogram plot to analysis the variable distributions over the whole dataset whether it's symmetric or not.

* Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups. They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outliers.

 Thus, for each numerical varibale in the given dataset, I used box plot to analyse the outliers and interquartile range including mean, median, maximum and minimum value.

Almost all columns are symmetric distributed and mean is nearly same with median for numerical columns.
